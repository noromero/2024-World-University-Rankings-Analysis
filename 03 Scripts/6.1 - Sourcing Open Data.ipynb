```python
# look at values in 'rank_order' and 'scores_overall_rank' by year
thewur_full.loc[thewur_full['year'].isin(range(2017,2025)),['year','rank_order','scores_overall_rank']]
```

`'rank_order'` has a different formatting between the years 2011-2015 and 2016-2024. For the latter years, an extra 0 was added to the end of a ranking to make rank of 1 have a rank order of 10, 2 -> 20, etc. That will be removed to have a consistent rank order

```python
# updating values from 2017-2024 that had an extra 0 at the end of rank order by dividing by 10
thewur_full.loc[thewur_full['year'].isin(range(2017,2025)),'rank_order'] = thewur_full.loc[thewur_full['year'].isin(range(2017,2025)),'rank_order'].div(10).astype(int)
thewur_full.loc[thewur_full['year'].isin(range(2017,2025)),'scores_overall_rank'] = thewur_full.loc[thewur_full['year'].isin(range(2017,2025)),'scores_overall_rank'].div(10).astype(int)
```

```python
# read csv with world bank data & list columns
worldbank_df = pd.read_csv(os.path.join(path, '02 Data', 'Original', 'WorldBank_EducationExpenditure_GDP.csv'),header=2)
worldbank_df.columns
```

```python
# check stats for new df
worldbank_df.describe()
```

```python
# create subset of data with only top200 schools
top200 = thewur_full[thewur_full['rank_order']<201]
```

```python
# iterate through list of countries in thewur df and return whether they are in wb df
ct=0
for country in thewur_full['location'].unique():
    if country not in (worldbank_df['Country Name'].unique()):
        ct+=1
        print(ct, '\t',country in (worldbank_df['Country Name'].unique()), '\t',country)
```

```python
# update country names to match across dataframes

worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Egypt, Arab Rep.','Egypt')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Macao SAR, China','Macao')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Turkiye','Turkey')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Hong Kong SAR, China','Hong Kong')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Korea, Rep.','South Korea')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Iran, Islamic Rep.','Iran')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Czechia','Czech Republic')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Slovak Republic','Slovakia')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Venezuela, RB','Venezuela')
worldbank_df['Country Name']=worldbank_df['Country Name'].replace('Viet Nam','Vietnam')
```

```python
# iterate through list of countries in thewur df and return whether they are in wb df
ct=0
for country in thewur_full['location'].unique():
    if country not in (worldbank_df['Country Name'].unique()):
        ct+=1
        print(ct, '\t',country in (worldbank_df['Country Name'].unique()), '\t',country)
```

```python
# number of countries with no world bank data
thewur_full[thewur_full['location'].isin(['Taiwan','Palestine','Northern Cyprus'])].describe()
```

```python
# create subset of world bank data with relevant years
wb_reduced = worldbank_df[['Country Name','2011','2012','2013','2014',
                    '2015','2016','2017','2018','2019','2020',
                    '2021','2022','2023']]


# new df to hold new data frame
wb_transformed = pd.DataFrame(columns=['location','year','gdp%'])

location_list = []
year_list = []
gdp_list = []

# iterate over each country and create a row for each year entry and the corresponding value
for country in wb_reduced['Country Name']:
    for year in range(2011,2024):
        location_list.append(country)
        year_list.append(year)
        gdp_list.append(wb_reduced[wb_reduced['Country Name']==country][str(year)].iloc[0])

# create empty rows for unlisted countries in wb df
for country in ['Taiwan','Palestine','Northern Cyprus']:
    for year in range(2011,2024):
        location_list.append(country)
        year_list.append(year)
        gdp_list.append(None)

# create empty row for each country to hold 2024 data
for country in wb_reduced['Country Name'].unique():
    location_list.append(country)
    year_list.append(2024)
    gdp_list.append(None)

# create row for 2024 for countries unlisted countries
for country in ['Taiwan','Palestine','Northern Cyprus']:
    location_list.append(country)
    year_list.append(2024)
    gdp_list.append(None)

wb_transformed['location'] = location_list
wb_transformed['year'] = year_list
wb_transformed['gdp%'] = gdp_list
```

```python
# check values in original df
wb_reduced.loc[wb_reduced['Country Name']=='Egypt']
```

```python
# check transposed values in new df
wb_transformed.loc[wb_transformed['location']=='Egypt']
```

```python
# merge df across country/year
wur_wb_merged = thewur_full.merge(wb_transformed, on = ['location','year'])
wur_wb_merged
```
